```python
import numpy as np
```

```python
angles = np.array([0.00000000, 0.08726646, 0.17453293, 0.26179939, 0.34906585, 
                   0.43633231, 0.52359878, 0.61086524, 0.69813170, 0.78539816, 
                   0.87266463, 0.95993109, 1.04719755, 1.13446401, 1.22173048, 
                   1.30899694, 1.39626340, 1.48352986, 1.57079633])
down = np.array([
    0.9755, 0.9755, 0.9755, 0.9755, 0.9750, 
    0.9750, 0.9750, 0.9725, 0.9725, 0.9700,
    0.9655, 0.9650, 0.9625, 0.9600, 0.9550,
    0.9455, 0.9400, 0.9350, 0.9300
])

up = np.array([
    0.9300, 0.9300, 0.9200, 0.9150, 0.9100,
    0.9000, 0.8900, 0.8850, 0.8750, 0.8700, 
    0.8650, 0.8600, 0.8600, 0.8550, 0.8600, 
    0.8650, 0.8650, 0.8700, 0.8800
])
```

```python
# Fit a polynomial to the upwelling data
up_fit = np.polyfit(angles, up, 4)

# Fit a polynomial to the downwelling data
down_fit = np.polyfit(angles, down, 4)
```

```python
# Visualize up 
import matplotlib.pyplot as plt

plt.plot(angles, down, label='down')

# Plot the fit
x = np.linspace(0, np.pi / 2, 100)
y = np.polyval(down_fit, x)
plt.plot(x, y, label='down fit', color='red')

plt.xlabel('Angle (radian)')
plt.ylabel('Downwelling Transmittance')
plt.title('Downwelling Transmittance vs. Angle')

# y-limit
plt.ylim(0.9, 1.0)

plt.show()
```

```python
# Visualize down
plt.plot(angles, up, label='up')

# Plot the fit
x = np.linspace(0, np.pi / 2, 100)
y = np.polyval(up_fit, x)
plt.plot(x, y, label='up fit', color='red')

plt.xlabel('Angle (radian)')
plt.ylabel('Upwelling Transmittance')
plt.title('Upwelling Transmittance vs. Angle')

# y-limit
plt.ylim(0.8, 1.0)

plt.show()
```

```python
# Print the coefficients
print('Downwelling fit coefficients:', down_fit)
print('Upwelling fit coefficients:', up_fit)
```
